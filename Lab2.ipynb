```python
import pandas as pd
import numpy as np
print("\n-----Task_A------\n")
df=pd.read_csv('student_data.csv', encoding='latin1')
print(df.head(),"\n")
print(df.tail(),"\n")
rows=df.shape[0]
print("No.of Rows=",rows)
colums=df.shape[1]
print("No.of columns=",colums,"\n")
if "Name" in df.columns:
    for names in df["Name"]:
        print(names)
print("\ncolumn datatypes:\n",df.dtypes)
print("\nDetailed info:",df.info())
#partB
print("\n-----Task_B------\n")
mean_att=df["Attendance"].mean()
print("\n averagr attendence of class:",mean_att)
print("\n average internal marks of class:",df["InternalMarks"].mean())
print("\n average external marks of class:",df["ExternalMarks"].mean())
print("\n Maximum internal marks:",df["InternalMarks"].max())
print("\n Minimum internal marks:",df["InternalMarks"].min())
print("\n Maximum external marks:",df["ExternalMarks"].max())
print("\n Minimum external marks:",df["ExternalMarks"].min())
print("\n Standard dev of internal marks:",df["InternalMarks"].std())
print("\n Summary Statistic:\n", df.describe())
#taskC
print("\n-----Task_c------\n")
df["TotalMarks"]=df[["InternalMarks","ExternalMarks"]].sum(axis=1)
df["Percentage"]=(df["TotalMarks"]/100)*100
print(df)
#taskD
print("\n-----Task_D------\n")
print("Students with the attendance greater than 90:\n")
print(df.loc[df["Attendance"] >= 90, "Name"])
print(df.loc[df["Attendance"] >= 90, "Name"])
print(df.loc[df["Gender"] == "F", "Name"])
print(df.loc[df["Gender"] == "M", "Name"])
print(df.loc[df["ExternalMarks"] <= 50, "Name"])
#paet e
print("\n-----Task_E------\n")

print(df.loc[df["TotalMarks"] == df["TotalMarks"].max(), "Name"])
print(df.loc[df["TotalMarks"] == df["TotalMarks"].min(), "Name"])
df.sort_values(by="TotalMarks", ascending=False, inplace=True)
print(df.head(3),"Name")
print(df.tail(3),"Name")
classavg=df["TotalMarks"].mean()
count=(df["TotalMarks"]>classavg).sum()
print("No of students scored above class avg:",count)
#Tasj=k f\
print("\n-----Task_F------\n")
print("Class average of males:", df.loc[df["Gender"] == "M", "TotalMarks"].mean())
print("Class average of females:", df.loc[df["Gender"] == "F", "TotalMarks"].mean())
M_att=df.loc[df["Gender"] == "M", "Attendance"].mean()
F_att=df.loc[df["Gender"] == "F", "Attendance"].mean()
if F_att>M_att:
    print("female have more attendence")
else:
    print("male have more attendence")
print("\n-----Task_G------\n")
import matplotlib.pyplot as plt
plt.bar(df["Name"], df["TotalMarks"], color="skyblue")
plt.xlabel("Students")
plt.ylabel("Total Marks")
plt.title("Total Marks of Students")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("\n")
plt.hist(df["Attendance"], bins=10, color="lightgreen", edgecolor="black")
plt.xlabel("Attendance")
plt.ylabel("Number of Students")
plt.title("Distribution of Attendance")
plt.show()
print("\n")
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue'])
plt.title('Gender Distribution of Students')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
print("\n")
plt.scatter(df["Attendance"], df["TotalMarks"], color="purple")
plt.xlabel("Attendance (%)")
plt.ylabel("Total Marks")
plt.title("Relationship between Attendance and Total Marks")
plt.show()
print("\n")
correlation = df["Attendance"].corr(df["TotalMarks"])
print("Correlation between Attendance and Total Marks:", correlation)
```
