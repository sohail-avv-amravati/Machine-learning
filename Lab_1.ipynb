```python
import numpy as np
import pandas as pd
print("S.Sohail")
print("24319")
print("CSE-B")
data={"Roll_NO":[101,102,103,104,105],"Name":["Ravi","Priya","Kiran","Anu","Sita"],"Marks":[85,92,78,88,95]}
df=pd.DataFrame(data)
print(df,"\n\n")
print(df.head(3),"\n\n\n")
average = df['Marks'].mean()
print(average)
higest=df.loc[df["Marks"].idxmax(),"Name"]
print("\n\n\n",higest)
lowest=df.loc[df["Marks"].idxmin(),"Name"]
print("\n\n\n",lowest)
conditions=[df["Marks"]>=90 ,df["Marks"]>=80,df["Marks"]<80]
grades=["A","B","C"]
df["Grades"]=np.select(conditions,grades, default="")
print(df)
```

```python
import pandas as pd
print("S.Sohail")
print("24319")
print("CSE-B\n")
data = {
    'Product': ['Laptop',"Mouse","Keyboard","Monitor"],
    'Price':[50000,500,1200,10000],
    'Quantity':[10,50,30,15]
}
df=pd.DataFrame(data)
print(df,"\n")
df['Total Value'] = df['Price'] * df['Quantity']
print("1 & 2. Updated DataFrame with 'Total Value':")
print(df,"\n")
highest_val_product = df.loc[df['Total Value'].idxmax(), 'Product']
print(f"3. Product with the highest total value: {highest_val_product}")
overall_value = df['Total Value'].sum()
print(f"\n4. Overall inventory value: Rs. {overall_value:,}")
```

```python
import pandas as pd
from sklearn.datasets import load_iris
print("S.Sohail")
print("24319")
print("CSE-B\n")
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
species_map = {i: name for i, name in enumerate(iris.target_names)}
df['species'] = df['species'].map(species_map)
print("\n2. First five records:")
print(df.head())
print("\n" + "="*50 + "\n")
rows, cols = df.shape
print(f"3. Number of Rows: {rows}")
print(f"   Number of Columns: {cols}")
print("\n" + "="*50 + "\n")
print("4. Column names:")
print(list(df.columns))
print("\n" + "="*50 + "\n")
print("5. Missing values per column:")
print(df.isnull().sum())
print("\n" + "="*50 + "\n")
print("6. Summary statistics:")
print(df.describe())
print("\n" + "="*50 + "\n")
print("7. Number of samples per class:")
print(df['species'].value_counts())

```
